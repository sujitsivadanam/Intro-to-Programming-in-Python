```
#1
def day_of_the_week(num):
  if num == 1:
    print('The 1st day of the week is Wednesday.')
  if num == 2:
    print('The 2nd day of the week is Thursday.')
  if num == 3:
    print('The 3rd day of the week is Friday.')
  if num == 4:
    print('The 4th day of the week is Saturday.')
  if num == 5:
    print('The 5th day of the week is Sunday.')
  if num == 6:
    print('The 6th day of the week is Monday.')
  if num == 7:
    print('The 7th day of the week is Tuesday.')

day_of_the_week(6)
```
